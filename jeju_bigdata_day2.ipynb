```python
# 문제 해결 방법 구상하기
# 질문 : 제주에서 가장 더웠던 날은 언제였을까? 기온은 몇 도 였을까?

f = open('jeju.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)

max_temp = -100
max_date = ''

for row in data:
  if row[-1]:
    row[-1] = float(row[-1])
  else:                          # 값이 없는 경우 0 으로 대체 @@중요함@@
    row[-1] = 0
  # print(row)
  if max_temp < row[-1]:
    max_temp = row[-1]
    max_date = row[0].replace('\t','')

f.close()

print(f"제주에서 가장 더웠던 날 : {max_date}, 그 날의 기온 : {max_temp} °")
```

```python
# 연도별 평균 기온을 선 그래프로 시각화
import csv
import matplotlib.pyplot as plt

f = open('jeju.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)

temp_avg_1980 = []
temp_avg_1990 = []
temp_avg_2000 = []
temp_avg_2010 = []
temp_avg_2020 = []

for row in data:
  if row[2]:
    row[2] = float(row[2])
  else:                          # 값이 없는 경우 0 으로 대체 @@중요함@@
    row[2] = 0
  if '1980' in row[0]:
    temp_avg_1980.append(row[2])
  if '1990' in row[0]:
    temp_avg_1990.append(row[2])
  if '2000' in row[0]:
    temp_avg_2000.append(row[2])
  if '2010' in row[0]:
    temp_avg_2010.append(row[2])
  if '2020' in row[0]:
    temp_avg_2020.append(row[2])

f.close()

plt.figure(figsize=(14,5))
plt.title('Average Temperature')
plt.plot(temp_avg_1980, lw=0.8, color='lightblue', label='1980')
plt.plot(temp_avg_1990, lw=0.8, color='skyblue', label='1990')
plt.plot(temp_avg_2000, lw=0.8, color='royalblue', label='2000')
plt.plot(temp_avg_2010, lw=0.8, color='blue', label='2010')
plt.plot(temp_avg_2020, lw=0.8, color='navy', label='2020')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

```python
# 우리 동네 연령별 인구 데이터 분석
import csv
import matplotlib.pyplot as plt

f = open('age.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)

name = input('동네 이름을 입력하세요 : ')

for row in data:
  if name in row[0]:
    result = []
    name_searched = ''
    for i in row[3:]:
      result.append(int(i.replace(',','')))
      name_searched = row[0]
    print(name_searched + ' 연령별 인구수')
    plt.plot(result)
    plt.grid(True)
    plt.show()

f.close()
```
