```python
# '대정읍' 남녀 인구 시각화

import csv
import matplotlib.pyplot as plt

f = open('gender.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)

male = []
female = []
loc = ''

for row in data:
  if '대정읍' in row[0]:
    loc = row[0]
    for i in range(0,101):
      male.append(int(row[i+3].replace(',','')))
      female.append(int(row[i+106].replace(',','')))
    break

f.close()

print(loc)
plt.plot(male, color='skyblue')
plt.plot(female, color='pink')
plt.legend(['male','female'])
plt.show()
```

```python
# 남녀 인구 시각화 (수평 막대그래프 & 좌우 분리)

import csv
import matplotlib.pyplot as plt

f = open('gender.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)


loc_input = input('분석할 지역 이름을 입력하세요 : ')

for row in data:
  if loc_input in row[0]:
    male = []
    female = []
    loc = ''
    loc = row[0]
    for i in range(0,101):
      male.append(-int(row[i+3].replace(',','')))
      female.append(int(row[i+106].replace(',','')))
    print(loc)
    plt.barh(range(0,101), male, color='skyblue', alpha=0.8)
    plt.barh(range(0,101), female, color='pink', alpha=0.8)
    plt.legend(['male','female'])
    plt.show()

f.close()
```

```python
# 제주도는 여성의 비율이 더 높을까?

import csv
import matplotlib.pyplot as plt

f = open('gender.csv','r',encoding='cp949')
data = csv.reader(f)
header = next(data)

loc_input = input('분석할 지역 이름을 입력하세요 : ')

for row in data:
  if loc_input in row[0]:
    male = 0
    female = 0
    loc = ''
    loc = row[0]
    for i in range(0,101):
      male += int(row[i+3].replace(',',''))
      female += int(row[i+106].replace(',',''))
    print(loc)
    plt.pie([male, female], labels = ['male', 'female'], autopct='%0.2f%%', startangle=90)
    plt.show()

f.close()
```

```python
# pandas 에 데이터 로딩 해보기

import pandas as pd

# age.csv 에서 데이터를 로딩
df = pd.read_csv('age.csv', encoding='cp949')

loc_name = input('분석할 지역을 입력하세요 : ')

# 분석할 지역 선택 및 필터링
df_filter = df[df['행정구역'].str.contains(loc_name)]

if not df_filter.empty:
  print()
  print(f"'{loc_name}' 지역의 인구 데이터를 찾았습니다.")
  print()
  # 원하는 데이터만 추출
  age_columns = [f"2025년07월_계_{i}세" for i in range(100)]
  age_data = df_filter[age_columns]
  print(age_data)
  print()
  # age_data 를 csv 파일로 저장
  output_filename = f'{loc_name}_age.csv'
  age_data.to_csv(output_filename, index=False, encoding='cp949')
  print(f"'{output_filename}' 파일로 저장했습니다.")

else:
  print('해당 지역의 데이터가 없습니다.')
```

```python
# pandas 에 데이터 로딩 해보기

import pandas as pd

# age.csv 에서 데이터를 로딩
df = pd.read_csv('age.csv', encoding='cp949')

loc_name = input('분석할 지역을 입력하세요 : ')

# 분석할 지역 선택 및 필터링
df_filter = df[df['행정구역'].str.contains(loc_name)]

if not df_filter.empty:
  print()
  print(f"'{loc_name}' 지역의 인구 데이터를 찾았습니다.")
  print()
  # 원하는 데이터만 추출
  age_columns = [f"2025년07월_계_{i}세" for i in range(100)]
  age_data = df_filter.iloc[0][age_columns]

  age_data = pd.to_numeric(age_data) # 문자열을 정수형으로 변환 @@@
  age_data.plot()

else:
  print('해당 지역의 데이터가 없습니다.')
```
